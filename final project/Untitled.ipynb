```python
#changing of the objects datatype to date-time datatype
data['clear_date'] = pd.to_datetime(data['clear_date'])
data['posting_date'] = pd.to_datetime(data['posting_date']) 
data['due_in_date'] = pd.to_datetime(data['due_in_date'], format='%Y%m%d')
data['document_create_date.1'] = pd.to_datetime(data['document_create_date.1'], format='%Y%m%d')
data['document_create_date'] = pd.to_datetime(data['document_create_date.1'], format='%Y%m%d')
data.head()
```
